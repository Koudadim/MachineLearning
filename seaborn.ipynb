# Le package seaborn

Il s'agit d'un package créé sur matplotlib et pandas et qui permet de faire une analyse de données très poussée.

## Data visualisation

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
iris = pd.read_csv('iris.csv')
iris.head()
iris.columns
```

```python
iris.rename(columns = {'SepalLengthcm':'SepalLength'})
iris.rename(columns = {'SepalWidthCm':'SepalWidth'})
iris.rename(columns = {'PetalLengthCm':'PetalLength'}) 
iris.rename(columns = {'PetalWidthCm':'PetalWidth'}) 
iris = iris.drop(['Id'], axis = 1)
```

```python
plt.scatter(iris['SepalLengthCm'], iris['SepalWidthCm'])
```

```python
import seaborn as sns
```

```python
sns.pairplot(iris)
plt.show()
```

```python
# Visualisation de la repartition des trois variétés de fleurs

sns.pairplot(iris, hue = 'Species')
plt.show()
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic.columns
```

```python
titanic.drop([ 'class', 'who', 'adult_male', 'embark_town','alive', 'alone'], axis = 1, inplace = True)
titanic.dropna(axis = 0, inplace = True)
titanic.head()
```

```python
# Visualisation des données à l'aide de pairplot
sns.pairplot(titanic)
```

```python
# Représentation de variables catégorielles
sns.catplot(x = 'pclass', y = 'age', data = titanic)
```

```python
# Repartition des sexes selon les classe
sns.catplot(x = 'pclass', y = 'age', data = titanic, hue = 'sex')
```

```python
# Reprséentation par le boxplot

sns.boxplot(x = 'pclass', y = 'age', data = titanic)
```

```python
sns.boxplot(x = 'pclass', y = 'age', data = titanic, hue = 'sex')
```

```python
# Visualisation de la distribution de la variable 'fare' (tarif)
sns.distplot(titanic['fare'])
```

```python
sns.jointplot('age', 'fare', data = titanic) 
```

```python
sns.jointplot(titanic['age'], titanic['fare'], data = titanic, kind = 'kde') 
plt.show()
```

```python
sns.jointplot('age', 'fare', data = titanic, kind = 'hex') # hexagonal
```

```python
# Visualisation de la matrice de correlation
sns.heatmap(titanic.corr())
```
